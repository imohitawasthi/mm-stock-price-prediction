```python
# Imports
import pickle

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from datetime import datetime
```

```python
DATA_FRAME_PATH = '/mnt/e/HOME/AI/model_stock_predictions/data/processed/processed_correlation.csv'
DATA_FRAME_PATH_PICKLE = '/mnt/e/HOME/AI/model_stock_predictions/data/processed/processed_correlation.pickle'

FINAL_DATA_FRAME_PATH = '/mnt/e/HOME/AI/model_stock_predictions/data/processed/completed_frame.csv'
```

```python
data_frame = data_frame = pd.read_csv(DATA_FRAME_PATH)
```

```python
data_frame
```

```python
predictions = pd.DataFrame()

data_frame = data_frame.to_dict('index')
```

```python
prediction = []

for key in data_frame.keys():    
    
    data = data_frame[key]
        
    symbol = data['Symbol']
    cost_correlation = round(data['CostCorrelation'], 2)
    cost_current = round(data['Current'], 2)
    cost_average = round(data['Average'], 2)
    change_percentage = round(data['MaxChangePercentage'], 2)
    average_predicted = round(data['PredictedAverage1Year'], 2)
    cost_predicted = round(data['PredictedClose1Year'], 2)
            
    predicted_score = (
        (1 if cost_correlation > 70 else 0)
        + 
        (1 if abs(cost_current/cost_average) > 0 and abs(cost_current/cost_average) < 3 else 0)
        +
        (1 if change_percentage > 10 and change_percentage < 50 else 0)
        +
        (10 * ((average_predicted - cost_average)/cost_average))
        +
        (10 * ((cost_predicted - cost_current)/cost_current))
    )
    
    predictions = predictions.append({
        'symbol': symbol,
        'cost_correlation': cost_correlation,
        'cost_current': cost_current,
        'cost_average': cost_average,
        'change_percentage': change_percentage,
        'average_predicted': average_predicted,
        'cost_predicted': cost_predicted,
        'predicted_score': predicted_score
    }, ignore_index=True)
    

predictions
```

```python
predictions.to_csv(FINAL_DATA_FRAME_PATH)
```
